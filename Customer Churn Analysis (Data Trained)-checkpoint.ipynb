```python
def datapreparation(filepath):
    churn=pd.read_csv(filepath)
    churn.drop(['customerID'],axis=1,inplace=True)
    churn.TotalCharges=churn.TotalCharges.replace(' ',np.nan)
    churn.TotalCharges.fillna(0,inplace=True)
    churn.TotalCharges=churn.TotalCharges.astype(float)
    cols1=['Partner','Dependents','PaperlessBilling','Churn','PhoneService']
    for col in cols1:
        churn[col]=churn[col].apply(lambda x:0 if x== 'No' else 1)
        churn.gender=churn.gender.apply(lambda x:0 if x=='Male' else 1)
        churn.MultipleLines=churn.MultipleLines.map({'No phone  service':0,'No':0,'Yes':1})
        cols2=['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']
        for col in cols2:
            churn[col]=churn[col].map({'No internet service':0,'No':0,'Yes':1})
            df=pd.get_dummies(churn,columns=['InternetService','Contract','PaymentMethod'],drop_first=True)
            return df
```

```python
df=datapreparation(filepath="customer_churn_alaysis.csv")
df.head()
```

```python
#Feature selection
x=df.drop('Churn',axis=1)
y=df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=42,stratify=y)

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
predrd=rf.predict(x_test)
print(confusion_matrix(y_test,predrd))
print(accuracy_score(y_test,predrd))
print(classification_report(y_test,predrd))
```

```python
#hyperparameter tuning
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators=[int(x) for x in np.linspace(start=40,stop=150,num=15)]
max_depth=[int(x) for x in np.linspace(40,150,num=15)]
```

```python
param_dist={'n_estimators':n_estimators,'max_depth':max_depth}
```

```python
rf_tuned=RandomForestClassifier(random_state=42)
```

```python
rf_cv=RandomizedSearchCV(estimator=rf_tuned,param_distributions=param_dist,cv=5,random_state=42)
```

```python
rf_cv.fit(x_train,y_train)
```

```python
rf_cv.best_score_
```

```python
rf_cv.best_params_
```

```python
rf_best=RandomForestClassifier(max_depth=134,n_estimators=71,random_state=42)
```

```python
rf_best.fit(x_train,y_train)
predy=rf_best.predict(x_test)
print(accuracy_score(y_test,predy))
print(confusion_matrix(y_test,predy))
print(classification_report(y_test,predy))
```
