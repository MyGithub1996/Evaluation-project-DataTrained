```python
#preparing for data modeling
x=df.drop(["Next_Tmax","Next_Tmin"],axis=1)
y=df[["Next_Tmax","Next_Tmin"]]
```

```python
#checking the shape of x data
print(x.shape)
```

```python
print(y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
best_rstate=0
accu=0
for i in range(30,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=i)
    mod=LinearRegression()
    mod.fit(x_train,y_train)
    predmod=mod.predict(x_test)
    tempaccu=r2_score(y_test,predmod)
    if tempaccu>accu:
        accu=tempaccu
        best_rstate=i
print(f'best accuracy={accu*100} found on random_state={best_rstate}')
print(mean_squared_error(y_test,predmod))
print(mean_absolute_error(y_test,predmod))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=43)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
prdrm=rf.predict(x_test)
r2score=r2_score(y_test,prdrm)
cvsscore=cross_val_score(RandomForestRegressor(),x_train,y_train,cv=5).mean()
print(f'accuracy={r2score*100},cross_val_score={cvsscore*100} & difference={(r2score*100)-(cvsscore*100)}')
print(mean_squared_error(y_test,prdrm))
print(mean_absolute_error(y_test,prdrm))
```

```python
#RandomForestRegressor is orking best amongst all others
#saving the model
import pickle
filename="Temperature_Prediction.pkl"
pickle.dump(rf,open(filename,"wb"))
```
