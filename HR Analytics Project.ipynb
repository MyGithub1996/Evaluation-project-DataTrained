### Create Testing and training datasets and perform cleaning

```python
y=df.Attrition
df.drop(['Attrition'],axis=1,inplace=True)
```

```python
df
```

```python
dummies=pd.get_dummies(df,columns=['BusinessTravel','Department','EducationField','JobRole','Gender','MaritalStatus'],
                       prefix='OH')
```

```python
dummies
```

```python
#normalising the data
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
X=Scaler.fit_transform(dummies)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0,shuffle=True)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Train and Evaluate a logistic regression classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
lg=LogisticRegression(max_iter=1000,C=100)
lg.fit(X_train,y_train)
predlg=lg.predict(X_test)
pred_lg=lg.predict_proba(X_test)
print(confusion_matrix(y_test,predlg))
print(classification_report(y_test,predlg))
print(accuracy_score(y_test,predlg))
```

```python
cm=confusion_matrix(y_test,predlg)
sns.heatmap(cm,annot=True)
plt.show()
```
